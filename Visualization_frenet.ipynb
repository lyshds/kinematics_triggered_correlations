```python
def plot_prediction_at_time(timestamp: float):
    
    plt.rcParams["figure.figsize"] = (20,10)
    plots = data.plot.line(y=['u_acceleration', 'u_steering_angle', 'Velocity', 'Eta', 'Theta'], subplots=True,)
    plt.suptitle("Prediction at timestamp:" + str(timestamp), fontsize=16)
    prediction_u = data.loc[timestamp]['pred_control']
    prediction_states = data.loc[timestamp]['pred_states']
    
    pred_time = np.arange(timestamp, timestamp + .2 *10, 0.2)
    
    plots[0].plot(pred_time, prediction_u[0], "r--", label='prediction')
    plots[1].plot(pred_time, prediction_u[1], "r--", label='prediction')
    plots[2].plot(pred_time, prediction_states[:, 3], "r--", label='prediction')
    plots[3].plot(pred_time, prediction_states[:, 5], "r--", label='prediction')
    plots[4].plot(pred_time, prediction_states[:, 6], "r--", label='prediction')
    
    [plots[i].hlines(0, timestamp -5, timestamp +5)for i in range(5)]
    
    
    # Zooming to prediction timestamp
    plt.xlim([timestamp-0.4, timestamp+2.4])
    # Update legend in every plot
    [plots[i].legend() for i in range(5)]
    plt.savefig('prediction_zoom.pdf') 
    return plots
```
